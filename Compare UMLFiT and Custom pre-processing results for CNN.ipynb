```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from bokeh.plotting import figure, output_notebook, show

output_notebook()
```

```python
ulmfit_results = pd.read_pickle('inspire_data_with_reference_fractions_and_num_references/predictions_ulmfit_preprocessing.pkl')
custom_results = pd.read_pickle('inspire_data_with_reference_fractions_and_num_references/predictions_custom_preprocessing.pkl')
```

```python
ulmfit_lc_pr = ulmfit_results[(ulmfit_results.label == 2) & (ulmfit_results.pred == 0)]
ulmfit_lc_pn = ulmfit_results[(ulmfit_results.label == 2) & (ulmfit_results.pred == 1)]
ulmfit_lc_pc = ulmfit_results[(ulmfit_results.label == 2) & (ulmfit_results.pred == 2)]
ulmfit_ln_pr = ulmfit_results[(ulmfit_results.label == 1) & (ulmfit_results.pred == 0)]
ulmfit_ln_pn = ulmfit_results[(ulmfit_results.label == 1) & (ulmfit_results.pred == 1)]
ulmfit_ln_pc = ulmfit_results[(ulmfit_results.label == 1) & (ulmfit_results.pred == 2)]
ulmfit_lr_pr = ulmfit_results[(ulmfit_results.label == 0) & (ulmfit_results.pred == 0)]
ulmfit_lr_pn = ulmfit_results[(ulmfit_results.label == 0) & (ulmfit_results.pred == 1)]
ulmfit_lr_pc = ulmfit_results[(ulmfit_results.label == 0) & (ulmfit_results.pred == 2)]
```

```python
x_lbls = ['Core1', 'Noncore1', 'Core2', 'Noncore2', 'Total1', 'Total2']

plt.figure(figsize=(16, 16))

for i in range(6):
    plt.subplot(9,6,i+1)
    plt.hist([ref[i] for ref in ulmfit_lr_pr.ref_fractions], bins=20)
    if i == 0:
        plt.ylabel('R-R', rotation=90)
        plt.title('ULMFiT preprocessing')

for i in range(6):
    plt.subplot(9,6,6+i+1)
    plt.hist([ref[i] for ref in ulmfit_lr_pn.ref_fractions], bins=20)
    if i == 0:
        plt.ylabel('R-NC', rotation=45)

for i in range(6):
    plt.subplot(9,6,12+i+1)
    plt.hist([ref[i] for ref in ulmfit_lr_pc.ref_fractions], bins=20)
    if i == 0:
        plt.ylabel('R-C', rotation=45)

for i in range(6):
    plt.subplot(9,6,18+i+1)
    plt.hist([ref[i] for ref in ulmfit_ln_pr.ref_fractions], bins=20)
    if i == 0:
        plt.ylabel('NC-R', rotation=45)

for i in range(6):
    plt.subplot(9,6,24+i+1)
    plt.hist([ref[i] for ref in ulmfit_ln_pn.ref_fractions], bins=20)
    if i == 0:
        plt.ylabel('NC-NC', rotation=45)

for i in range(6):
    plt.subplot(9,6,30+i+1)
    plt.hist([ref[i] for ref in ulmfit_ln_pc.ref_fractions], bins=20)
    if i == 0:
        plt.ylabel('NC-C', rotation=45)

for i in range(6):
    plt.subplot(9,6,36+i+1)
    plt.hist([ref[i] for ref in ulmfit_lc_pr.ref_fractions], bins=20)
    if i == 0:
        plt.ylabel('C-R', rotation=45)

for i in range(6):
    plt.subplot(9,6,42+i+1)
    plt.hist([ref[i] for ref in ulmfit_lc_pn.ref_fractions], bins=20)
    if i == 0:
        plt.ylabel('C-NC', rotation=45)

for i in range(6):
    plt.subplot(9,6,48+i+1)
    plt.hist([ref[i] for ref in ulmfit_lc_pc.ref_fractions], bins=20)
    if i == 0:
        plt.ylabel('C-C', rotation=45)
    plt.xlabel(x_lbls[i])
```
